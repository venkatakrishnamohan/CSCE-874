```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('association_data.csv')
```

```python
pd.options.display.max_rows = 999
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0'],inplace=True,axis=1)
```

```python
from sklearn import decomposition
```

```python
pca = decomposition.PCA(n_components=1)
```

```python
ana = ['Government School For Disabled (Status A(1)/NA(2))',
       'Govt Engineering College (Status A(1)/NA(2))',
       'Govt Management Institute (Status A(1)/NA(2))',
       'Govt Medicine College (Status A(1)/NA(2))',
       'Govt Middle  School (Status A(1)/NA(2))',
       'Govt Primary School (Status A(1)/NA(2))',
       'Govt Secondary School (Status A(1)/NA(2))',
       'Govt Senior Secondary School (Status A(1)/NA(2))',
         'Private Engineering College (Status A(1)/NA(2))',
       'Private Management Institute (Status A(1)/NA(2))',
       'Private Medicine  College (Status A(1)/NA(2))',
       'Private Middle  School (Status A(1)/NA(2))',
       'Private Primary School (Status A(1)/NA(2))',
       'Private School For Disabled ( Status A(1)/NA(2))',
       'Private Secondary School (Status A(1)/NA(2))',
       'Private Senior Secondary School (Status A(1)/NA(2))',
       'Public Library (Status A(1)/NA(2))']
```

```python
pop = [ 'F_ILL', 'F_LIT',
     'MAINWORK_P', 'MAIN_AL_P', 'MAIN_CL_P',
       'MAIN_HH_P', 'MARGWORK_F', 'MARGWORK_M', 'MARGWORK_P', 'MARG_AL_P',
       'MARG_CL_P', 'MARG_HH_P', 'M_ILL', 'M_LIT',
        'NON_WORK_F',
       'NON_WORK_M', 'NON_WORK_P', 'No_HH', 'P_ILL', 'P_LIT', 'P_SC', 'P_ST',
         'TOT_F', 'TOT_M', 'TOT_P',
       'TOT_WORK_F', 'TOT_WORK_M', 'TOT_WORK_P', 
       ]
infra_area = ['Dispensary (Numbers)','Hospiltal Alternative Medicine (Numbers)',
       'Hospital Allopathic (Numbers)','Maternity And Child Welfare Centre (Numbers)','Veterinary Hospital (Numbers)']
social_housing = ['Availing banking','Has Mobile phone only', 'Has TV', 'Having toilet facility', 'Total good houses']
distance = ['District Head Quarter\xa0 (Distance in km)','Sub District Head Quarter (Distance in km)']
```

```python
infra_num = ['NUMbanks', 'NUMfirestations',
       'NUMschools', 'NUMbusstations', 'NUMParks', 'NUMcollege',
       'NUMgasstations', 'NUMgovt.offices', 'NUMhospitals',
       'NUMPolicestations', 'NUMsports', 'NUMtoursim', 'NUMPlacesofworship']
infra_dist = ['bank_distance', 'Bus_stop_distance', 'College_distance',
       'Fire_station_distance', 'gas_station_distance', 'govt.office_distance',
       'Hospitals_distance', 'parks_distance', 'places_of_worship_distance',
       'police_stations_distance', 'Schools_distance', 'sports_distance',
       'tourism_distance']
```

```python
pca.fit(df[infra_dist])
X = pca.transform(df[infra_dist])
```

```python
X
```

```python
X=(X-X.min())/(X.max()-X.min())
```

```python
df['Infra Distance'] = X
```

```python
all_attrs = ana+pop+infra_area+infra_dist+infra_num+social_housing+distance
```

```python
df.drop(all_attrs,inplace=True,axis=1)
```

```python
df.head()
```

```python
df['Infra Area'] = pd.qcut(df['Infra Area'].rank(method='first').values,3,labels = ['poor','better','good'],duplicates='drop')
```

```python
df.to_csv('association_pca.csv')
```
