```python
# Read the vocabulary related to social unrest
unrest_data = pd.read_excel('SocialUnrestVocabulary copy.xlsx')
```

```python
#Get only the english words from the excel file
unrest_data_eng = unrest_data['English 1']
```

```python
import glob
import os
```

```python
import csv
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
#Main program which iterates through all the files in the provided folder
for filename in glob.glob("../News_articles/Januaury/*.txt"):
    print(filename)
    f =open(filename)
    # read the contents of the file
    raw = f.read()
    tokens = nltk.word_tokenize(raw)
    text = nltk.Text(tokens)
    # Remove single character words (a, i)
    words = [word for word in text if len(word) > 1]
    # Fix the spellings of the words
    words = [spell(word) for word in words]
    # Convert all words to lower case
    words = [word.lower() for word in words if word.isalpha()]
    # remove the stop words from the tokens (an, the, it, and)
    words = [word for word in words if word not in stop_words]
    # Lemmatize all words (all verb forms are condensed to their root word)
    words = [wordnet_lemmatizer.lemmatize(word) for word in words]
    words = [wordnet_lemmatizer.lemmatize(word,'v') for word in words]
    # Get the count of all words in the file
    fdist1 = FreqDist(words)
    a={}
    count=0
    # Iterate over the words from unrest vocabulary
    for word in unrest_data_eng:
        a[word] = fdist1[word]
        # If the unrest word is found then increment the count by 1
        count = count+fdist1[word]
        # If the number of unrest words in the file are more than 5 then save them in to a separate folder and their counts 
        #into another folder
        if count>5:
            c = csv.writer(open("Count/"+os.path.basename(f.name)+"_count.csv",'w'))
            for key,val in a.items():
                c.writerow([key,val])
            f1 = open("Unrest_articles/"+os.path.basename(f.name),'w')
            f1.write(raw)
        
```
