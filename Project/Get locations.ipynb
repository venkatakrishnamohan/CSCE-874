```python
# Main part of the program
i=0
# Iterate over all unrest related files
for filename in glob.glob("subset_unrest2/*.txt"):
    try:
        print(filename)
        # Open and read the file
        f = open(filename)
        raw = f.read()
        city = nltk.word_tokenize(raw.splitlines()[3])[0]
        # Get the places object from geograpy
        places = geograpy.get_place_context(text = raw)
        if len(places.countries)==0:
            country = 'India'
        elif 'India' in places.countries:
            country = 'India'
        else:
            continue
        # Get the city names
        cities = places.cities
        # get the count of different cities in the article
        cities_ct = Counter(cities)
        # get the most common city name from the article
        if len(cities_ct)>0:
            most_common_city = cities_ct.most_common(1)[0][0]
        else:
            most_common_city=""
        # get the other places which might be provided by the geograpy package. This might contain some names of people. 
        other = places.other
        other_ct = Counter(other)
        # get the 3 most common names from the other places list reported by geograpy
        if len(other_ct)>0:
            most_common_1_other = other_ct.most_common(1)[0][0]
        else:
            most_common_1_other = ""
        if len(other_ct)>1:
            most_common_2_other = other_ct.most_common(2)[1][0]
        else:
            most_common_2_other = ""
        if len(other_ct)>2:
            most_common_3_other = other_ct.most_common(3)[2][0]
        else:
            most_common_3_other = ""
        # Prepare a location string by combining the 3 most common names from other field, city name, country.
        loc = most_common_1_other + ","+ most_common_2_other+","+most_common_3_other+","+ most_common_city+","+ city+","+country
        # geocode the location by providing this location string to google maps API
        location = g.geocode(loc,timeout=15)
        # If the google maps doesn't return anything then make the location string as just the city name.
        if location == None:
           loc = city+","+country
           location = g.geocode(loc,timeout=15) 
        i=i+1
        #print(location.latitude,location.longitude)
        # Save the latitude and longitude positions along with the article name in a dataframe.
        df.loc[i] = [location.latitude,location.longitude,os.path.basename(f.name)]
    except:
        continue
df.to_csv('event_locations_text_mining2.csv')
```
